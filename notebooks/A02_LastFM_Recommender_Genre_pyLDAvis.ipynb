```python
data = tag_data.tags
n_topics = 6
seed = 42
# ngram_range = (1,3)
# ncols = 2

# sw = set(STOPWORDS)

# adds = ['rock','pop','hardcore','vocalist','male','female']
# for word in adds:
#     sw.add(word)
```

### pyLDAvis

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 1.0, 
                                min_df = 1)
dtm_tf = tf_vectorizer.fit_transform(data)
print(dtm_tf.shape)
```

```python
lda_tf = LatentDirichletAllocation(n_topics=n_topics, random_state=seed)
lda_tf.fit(dtm_tf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer,mds='pcoa')
```
